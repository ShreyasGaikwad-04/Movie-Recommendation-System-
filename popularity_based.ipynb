## Popularity Based Recommender System ##
- Recommend top-rated or most-watched movies.
- Same Recommendations for all users.

```python
ratings = pd.read_csv("ratings.csv")
movies = pd.read_csv("movies.csv")
```

```python
print(ratings.head())
```

```python
print(movies.head())
```

```python
n_u = ratings["userId"].nunique()
n_m = movies["movieId"].nunique()

print(f'No. of Unique Users: {n_u}')
print(f'Total No. of Movies: {n_m}')
```

```python
# calculate avg rating and count of ratings per movie
movie_stats = ratings.groupby("movieId").agg(
    avg_rating=("rating", "mean"),
    num_ratings=("rating", "count")
)

# Filter Movies (the movie must be rated by atleast 10% of total users)
popular_movies = movie_stats[movie_stats["num_ratings"] >= (0.1 * n_u)]

# merge with movies dataset to get titles
popular_movies_with_titles = movies.merge(popular_movies, on="movieId")

# select top 10 by avg rating (sort once only)
top_movies_with_titles = popular_movies_with_titles.sort_values(
    "avg_rating", ascending=False
).head(10).reset_index(drop=True)
```
